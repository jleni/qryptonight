## Modeling

$\mu_t = \frac{D}{H_r}$

$Var_t = \frac{D \cdot (D-1)}{H_r^2}$


```python
# evenly sampled time at 200ms intervals
Hr = 35

D = np.arange(0., 10000., 10.)
mean = D/Hr
sd = (D*(D-1)/Hr**2)**0.5

lb = mean - sd
ub = mean + sd

# red dashes, blue squares and green triangles
plt.plot(D, mean, 'r--', ub, 'b--', lb, 'b--')
plt.ylim(0, 200)
plt.show()
```
